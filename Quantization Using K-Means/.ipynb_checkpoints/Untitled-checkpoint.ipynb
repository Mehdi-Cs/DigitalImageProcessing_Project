```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#############################################ORIGINAL IMAGE#########################################
img = cv2.imread(r'./2271214_c9ab_2.jpg')
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

```python
tmp = image.reshape((-1, 3))
tmp = np.float32(tmp)
crit = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
```

```python
k = 4
ret, label, center = cv.kmeans(tmp, k, None, crit, 10, cv.KMEANS_RANDOM_CENTERS)
center = cv.convertScaleAbs(center)
pic = center[label.flatten()]
result = pic.reshape((image.shape))
plt.imshow(result)
plt.show()
```
