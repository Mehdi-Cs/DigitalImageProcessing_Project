```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread(r'./image-20160712-9264-zt66ib.jpg')
```

```python
#####################################ORIGINAL IMAGE#############################################
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
height,width=image.shape[:2]
plt.imshow(image)
plt.show()
```

```python
#####################################Scalling 2x#############################################

scaledImage=cv2.resize(image,(2*width,2*height) )
plt.imshow(scaledImage)
plt.show()
```

```python
#####################################Ratating 45 Degrees#############################################
center = (width / 2, height / 2)
rotationMatrix = cv2.getRotationMatrix2D(center, 45, 1)
rotatedImage = cv2.warpAffine(image, rotationMatrix, (width, height))
plt.imshow(rotatedImage)
plt.show()
```

```python
#####################################Translating#############################################
translationMatrix = np.float32([ [1,0,70], [0,1,150] ])
translatedImage = cv2.warpAffine(image, translationMatrix, (width, height))
plt.imshow(translatedImage)
plt.show()
```

```python
#####################################Shearing#############################################
M2 = np.float32([[1, 0, 0], [0.2, 1, 0]])
M2[0,2] = -M2[0,1] * width/2
M2[1,2] = -M2[1,0] * height/2
shearedImage = cv2.warpAffine(image, M2, (width, height))
plt.imshow(shearedImage)
plt.show()
```
