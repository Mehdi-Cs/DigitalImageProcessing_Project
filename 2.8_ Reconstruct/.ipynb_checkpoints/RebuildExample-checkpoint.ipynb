```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread(r'./black-and-white-globe-png-3.png')
```

```python
#####################################ORIGINAL IMAGE#############################################
plt.figure(figsize=(6,6))
plt.imshow(img)
plt.show()
```

```python
#########################################Complement Image############################################
imageInvert = (255-img)
plt.figure(figsize=(6,6))
plt.imshow(imageInvert)
plt.show()
```

```python
##############################################Union Image#######################################################
imageUnion = np.maximum(np.array(img),np.array(imageInvert))
plt.figure(figsize=(6,6))
plt.imshow(imageUnion)
plt.show()
plt.close()
```
