```python
from numpy import fft
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
lena = cv2.imread('Lena.tif',cv2.IMREAD_GRAYSCALE)
plt.imshow(lena,cmap='gray')
plt.show()
plt.close()
```

```python
lena.shape
```

```python
I_fft = fft.fft2(lena)
I_phase = np.angle(I_fft)
plt.imshow(I_phase,cmap='gray')
plt.show()
plt.close()
```

```python
I_fft_recon =  np.exp(1j*I_phase)
I_ifft_recon = fft.ifft2(I_fft_recon)
I_ifft_recon = np.array(I_ifft_recon, dtype=np.float32)
plt.imshow(I_ifft_recon,cmap='gray')
plt.show()
plt.close()
```

```python
I_amp = abs(I_fft);
I_fft_recon = I_amp
I_ifft_recon = fft.ifft2(I_fft_recon)
I_ifft_recon = np.array(I_ifft_recon, dtype=np.float32)
center = np.fft.fftshift(I_ifft_recon)
plt.imshow(center,cmap='gray')
plt.show()
plt.close()
```
